```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# open image
img = cv2.imread('luise.jpg')

# convert to gray
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Laplacian
'''
Second derivation - edges
'''
laplace = cv2.Laplacian(gray, cv2.CV_64F, ksize=3)

# Convert to uint8
laplace = cv2.convertScaleAbs(laplace)

# Equalize the image of the laplacian filter
equalized_laplacian = cv2.equalizeHist(laplace)


# Show the original, the equalized image and their histograms
# In python we can calculate and show the histogram using only matplotlib
f, axs = plt.subplots(2,2,figsize=(15,15))
plt.figure(1)
plt.subplot(221)
plt.imshow(laplace, cmap='gray')
plt.subplot(222)
plt.hist(laplace.ravel(), 256, [0, 256])
plt.subplot(223)
plt.imshow(equalized_laplacian, cmap='gray')
plt.subplot(224)
plt.hist(equalized_laplacian.ravel(), 256, [0, 256])
plt.show()

#cv2.waitKey(0)
```
