```python
import cv2
import numpy as np
from matplotlib import pyplot as plt


# Open image
img = cv2.imread('luise.jpg')

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Histogram Equalization
dst = cv2.equalizeHist(gray)

# Calculate histograms of the original image and the equalized image using only openCV
original_hist = cv2.calcHist(gray, channels=[0], mask=None, histSize=[256], ranges=[0, 256])
equalized_hist = cv2.calcHist(dst, channels=[0], mask=None, histSize=[256], ranges=[0, 256])

# Show the original, the equalized image and their histograms
# In python we can calculate and show the histogram using only matplotlib
f, axs = plt.subplots(2,2,figsize=(15,15))
plt.figure(1)
plt.subplot(221)
plt.imshow(gray, cmap='gray')
plt.subplot(222)
plt.hist(gray.ravel(), 256, [0, 256])
plt.subplot(223)
plt.imshow(dst, cmap='gray')
plt.subplot(224)
plt.hist(dst.ravel(), 256, [0, 256])
plt.show()


#cv2.imshow('Source image', gray)
#cv2.imshow('Equalized Image', dst)

cv2.waitKey(0)
```
