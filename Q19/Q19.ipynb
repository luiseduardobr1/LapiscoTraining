```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# open image
img = cv2.imread('luise.jpg')

# convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Sobel operator - differentiation
Gx = cv2.Sobel(gray, dx = 1, dy = 0, ddepth=cv2.CV_64F, ksize = 3)
Gy = cv2.Sobel(gray, dx = 0, dy = 1, ddepth=cv2.CV_64F, ksize = 3)
sobel = (Gx**2 + Gy**2)**(1/2)

# Convert to uint8
sobel = cv2.convertScaleAbs(sobel)

# Show image and histogram
f, axs = plt.subplots(2,2,figsize=(15,15))
plt.figure(1)
plt.subplot(221)
plt.imshow(gray, cmap='gray')
plt.subplot(222)
plt.hist(gray.ravel(), 256, [0, 256])
plt.subplot(223)
plt.imshow(sobel, cmap='gray')
plt.subplot(224)
plt.hist(sobel.ravel(), 256, [0, 256])
plt.show()
```
